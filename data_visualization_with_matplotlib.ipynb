1. Plot am-based histogram to compare mpg (20 points)

```python
plt.style.use('ggplot')
data.groupby('am').mpg.hist(alpha=0.5, legend=True)
plt.xlabel('Miles per Gallon')
plt.ylabel('Histogram')
plt.title('Histogram of MPG by AM')
```

The histogram shows that cars labeled 1 in the AM (cars with manual transmission) have higher miles per gallon. The most frequently occuring mpg value for manual cars is greater than that of automatic cars.

2. Use scatterplot to plot mpg VS. hp (20 points)

```python
plt.figure()
plt.scatter(x=data.hp, y=data.mpg, marker='x')
plt.xlabel('Horse Power')
plt.ylabel('Miles per Gallon')
plt.title('MPG vs HP')
```

For this plot, I was interested in exploring the effect horse power has on miles per gallon, which is why hp is plotted on the x-axis and mpg is plotted on the y-axis. In general, we can see there is a strong negative relationship between the two variables. In other words, cars with higher horse powers get less miles per gallon. The relationship appears to be more exponential than linear.
